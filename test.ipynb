```python
import yfinance as yf
import pandas as pd
from time import sleep
import random

def fetch_financials(ticker):
    """
    Returns a dict of financial metrics for the given ticker.
    Includes: income statement, balance sheet, cash flow for the last 4 years.
    """
    try:
        stock = yf.Ticker(ticker)
        # Annual financials
        income = stock.financials
        balance = stock.balance_sheet
        cashflow = stock.cashflow

        # If data is empty, return None
        if income.empty:
            return None

        # Extract key metrics (example)
        revenue = income.loc['Total Revenue'] if 'Total Revenue' in income.index else None
        ebitda = income.loc['EBITDA'] if 'EBITDA' in income.index else None
        net_income = income.loc['Net Income'] if 'Net Income' in income.index else None
        total_assets = balance.loc['Total Assets'] if 'Total Assets' in balance.index else None
        total_liabilities = balance.loc['Total Liabilities'] if 'Total Liabilities' in balance.index else None
        operating_cf = cashflow.loc['Operating Cash Flow'] if 'Operating Cash Flow' in cashflow.index else None
        capex = cashflow.loc['Capital Expenditure'] if 'Capital Expenditure' in cashflow.index else None
        fcf = operating_cf + capex if operating_cf is not None and capex is not None else None

        # Convert to dict with years as keys
        data = {}
        for col in income.columns:
            year = col.year
            data[year] = {
                'revenue': revenue[col] if revenue is not None else None,
                'ebitda': ebitda[col] if ebitda is not None else None,
                'net_income': net_income[col] if net_income is not None else None,
                'total_assets': total_assets[col] if total_assets is not None else None,
                'total_liabilities': total_liabilities[col] if total_liabilities is not None else None,
                'operating_cf': operating_cf[col] if operating_cf is not None else None,
                'capex': capex[col] if capex is not None else None,
                'fcf': fcf[col] if fcf is not None else None,
            }
        return data
    except Exception as e:
        print(f"Error fetching financials for {ticker}: {e}")
        return None
    
msft_df = fetch_financials("MSFT")
print(msft_df)
```

```python
# !pip install edgartools
import re
from bs4 import BeautifulSoup
from edgar import *
import requests


set_identity("oscarchanly@gmail.com")

headers = {'User-Agent':'Dummy Company oscarchanly@gmail.com','Accept-Encoding':'gzip, deflate','Host':'www.sec.gov'}


def get_n_year_10K(ticker, n = 5): 

    """
    Extract 10 K filings in the last 5 years
    """

    # get 10-K filing from last n year 
    company = Company(ticker)

    filings_5y = company.get_filings(form="10-K").latest(n) 

   
    filings_list = [filing for filing in filings_5y]

    # print summary
    for filing in filings_list:
        print(f"{filing.filing_date}: {filing.form} - {filing.company}")

    return filings_list

filing_list_5y = get_n_year_10K("AAPL", 5)
```

```python

def normalize_text(text):
    """

    clean the text for pattern matching

    """
    # Replace non-breaking spaces
    text = text.replace('\u00A0', ' ')
    # # Replace curly apostrophe and other quotes with ASCII apostrophe
    text = text.replace('’', "'").replace('‘', "'").replace('`', "'")

    return text

def extract_mda_from_filing(filing):

    """
    Extract the MD&A (Item 7) text from a 10-K Filing object.
    """

    # Get the primary document's HTML
    url = filing.filing_url
    response = requests.get(url, headers = headers)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    text = soup.get_text(separator='\n')

    start_patterns = [
        r"item[ \t]*7\.?[ \t]*management's",
        r"item[ \t]*7\.?[ \t]*–[ \t]*management's",
    ]
    end_patterns = [
        r"item[ \t]*7a\.?[ \t]*quantitative",
        r"item[ \t]*8\.?[ \t]*financial\s+statements",
        r"item[ \t]*8\.?[ \t]*–[ \t]*financial\s+statements"
    ]
    lower_text = normalize_text(text.lower())
    
    start_idx = None
    for pat in start_patterns:
        match = re.search(pat, lower_text)
        if match:
            start_idx = match.start()
            break
    if start_idx is None:
        return None
    
    end_idx = len(text)
    for pat in end_patterns:
        match = re.search(pat, lower_text[start_idx:])
        if match:
            end_idx = start_idx + match.start()
            break
        
    
    mda_text = text[start_idx:end_idx].strip()
    mda_text = re.sub(r'^item\s*7\.?\s*', '', mda_text, flags=re.IGNORECASE).strip()
    print(url)
    return mda_text

# demostration 
print(extract_mda_from_filing(filing_list_5y[0]))
```

## Innovation Vocabulary Decay
**Definition:** 

Decline in the usage of forward‑looking words (e.g., “anticipate”, “expect”, “future”, “plan”) over time

**Approach:**

Use a predefined forward‑looking word list (e.g., from Loughran‑McDonald).

For each year, compute the frequency of these words (count / total words).

Fit a linear trend over the last 3–5 years; a negative slope indicates decay.


```python
from collections import Counter
import numpy as np

forward_words = {
    'anticipate', 'expect', 'believe', 'future', 'plan', 'potential',
    'opportunity', 'goal', 'objective', 'strategy', 'forecast', 'project',
    'outlook', 'target', 'aim', 'seek', 'hope', 'intend', 'look forward'
}

def compute_innovation_decay(mda_by_year):
    """
    Returns a dictionary year -> forward-looking word frequency,
    and the overall decay (slope) over the available years.
    """
    freq = {}
    for year, text in mda_by_year.items():
        words = re.findall(r'\b[a-zA-Z]+\b', text.lower())
        total_words = len(words)
        if total_words == 0:
            freq[year] = 0
            continue
        fwd_count = sum(1 for w in words if w in forward_words)
        freq[year] = fwd_count / total_words

    # Compute linear trend if at least 3 years
    years = sorted(freq.keys())
    if len(years) >= 3:
        x = np.array(years)
        y = np.array([freq[y] for y in years])
        slope = np.polyfit(x, y, 1)[0]  # negative slope = decay
    else:
        slope = None

    return freq, slope
```
